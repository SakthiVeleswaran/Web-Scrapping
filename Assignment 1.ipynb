# Web Scrapping

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'https://myanimelist.net/topanime.php'
```

```python
res = requests.get(url)
```

```python
res
```

```python
wave = BeautifulSoup(res.text)
```

```python
wave
```

```python
Anime = wave.select('.anime_ranking_h3 a')
```

```python
Anime
```

```python
Anime_title = []
```

```python
for item in Anime:
    Anime_title.append(item.text)
```

```python
Anime_title
```

```python
Episode = wave.select('.mt4')
```

```python
Episode
```

```python
Episode_title = []
```

```python
for item in Episode:
    Episode_title.append(item.text)
```

```python
Episode_title
```

```python
Score = wave.select('.on.score-label')
```

```python
Score
```

```python
Score_title = []
```

```python
for item in Score:
    Score_title.append(item.text)
```

```python
Score_title
```

```python
import pandas as cv
```

```python
av = cv.DataFrame(Anime_title,columns=['Anime'])
```

```python
av
```

```python
av["Episode"] = Episode
```

```python
av
```

```python
av["Score"] = Score
```

```python
av
```

```python
av.to_csv('av.csv')
```
